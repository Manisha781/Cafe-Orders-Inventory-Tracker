```python


                             #  Café Orders & Inventory Tracke

menu = {
    "tea": 120,
    "coffee": 200,
    "sandwich": 450,
    "cookie": 150,
    "cake_slice": 320
}

inventory = {
    "tea": 40,
    "coffee": 35,
    "sandwich": 20,
    "cookie": 50,
    "cake_slice": 25
}

orders = [
    {"tea": 2, "cookie": 1},
    {"coffee": 1, "sandwich": 2},
    {"tea": 3, "cake_slice": 1},
    {"coffee": 2, "cookie": 2, "sandwich": 1},
    {"tea": 1},
    {"coffee": 1, "cake_slice": 2}
]

print("Initial Inventory:", inventory)
print("Orders:", orders)
```

```python
processed_orders = []

for order in orders:
    cleaned_order = {}
    for item in order:
        if item in menu:
            if order[item] <= inventory[item]:
                cleaned_order[item] = order[item]
            else:
                cleaned_order[item] = inventory[item]
    processed_orders.append(cleaned_order)

print("Processed Orders:", processed_orders)
```

```python
order_totals = []

for order in processed_orders:
    total = 0
    for item in order:
        total += order[item] * menu[item]
    order_totals.append(total)

print("Order Totals (Before Discount):", order_totals)
```

```python
items_sold = {item: 0 for item in menu}

for order in processed_orders:
    for item in order:
        inventory[item] -= order[item]
        items_sold[item] += order[item]

print("Updated Inventory:", inventory)
print("Items Sold:", items_sold)
```

```python
low_stock = []

for item in inventory:
    if inventory[item] <= 5:
        low_stock.append(item)

print("Low Stock Items:", low_stock)
```

```python
order_totals_with_discount = []

for total in order_totals:
    if total > 1000:
        total = total - (0.10 * total)
    order_totals_with_discount.append(total)

print("Order Totals (With Discount):", order_totals_with_discount)
```

```python
order_points = []
total_points = 0

for total in order_totals_with_discount:
    points = int(total // 100)
    order_points.append(points)
    total_points += points

print("Loyalty Points per Order:", order_points)
print("Total Loyalty Points:", total_points)
```
